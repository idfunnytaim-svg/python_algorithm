```python
# 1. 팀장 번호 구하기 (산술 연산자)
# 1번부터 100번까지 학생이 있다.
# 1~10번은 1팀 
# 11~20번은 2팀 
# 21~30번은 3팀 
# ... 
# 각 팀은 10명씩이며 첫 번째 학생이 팀장이다.
# 학생 번호를 입력받아 해당 학생의 팀장 번호를 출력하는 프로그램을 작성하시오.
# 실행 예
# 학생 번호 : 8
# 팀장 번호 : 1
# 학생 번호 : 15
# 팀장 번호 : 11
```

```python
num = int(input("학생 번호 : "))

leader = ((num - 1) // 10) * 10 + 1

print("팀장 번호 :", leader)
```
