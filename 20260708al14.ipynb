```python
#판매액의 누적합
prices=[]
for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# prices의 누적합 계산
cumulative_sum = []
total = 0
for p in prices:
    total += p
    cumulative_sum.append(total)
```

```python
print("판매액의 누적합:", cumulative_sum)
```

```python
#판매액의 평균 누적합
avgs=[]
for i in range(len(cumulative_sum)):
    avg = cumulative_sum[i] / (i + 1)
    avgs.append(avg)
print("판매액의 평균 누적합:", avgs)
```

```python
# 1. 정규표현식 패턴

# ​

# 정규표현식의 패턴은 문자열의 규칙을 표현하는 식이다.

# ​

# 예를 들어 "010-1234-5678" 과 같은 전화번호를 찾고 싶다면 

# "010-"문자뒤에 숫자4개의 문자뒤에 "-"문자뒤에 숫자4개문자 

# 패턴을 정의할 수 있는데 이걸 정규표현식으로 나타내면 "010-\d{4}-\d{4}"와 같다.

# 패턴을 작성하기 위해서는 여러 가지 메타문자를 알아야 한다.

# ​

# ​

# 2. 자주 사용되는 정규표현식 메타문자

# ​

# 파이썬에서 정규표현식을 사용하여 문자열을 패턴을 찾을때는 re모듈을 사용한다.

# re모듈에 대해서는 메타문자 수업 이후 자세히 다룬다.

# ​

# 1) Raw String

# - r"" 로우 문자열(Raw String)은 문자열안에 백슬래시(\) 문자를 이스케이프 문자로 해석하지 않고 그대로 저장하는 문자열 표현이다. 정규표현식 
# 문법에 \가 자주 나오는데 매번 백슬래시를 "\\"표현하기 불편해서 r문자열을 사용
#"\\d" -> r"\d"
```

```python
print("Hello\nworld\\")
```

```python
print(r"Hello\nworld\\")
```

```python
# 2) 문자클래스 []
# - []안에 있는 문자 중 하나와 일치한다.
# - re모듈의 findall(정규표현식, text) 함수는 
# text문자열의 0부터 끝까지 정규표현식이 일치하는 문자열들을 반환한다.  
```

```python
import re

text = "apple banana cat"
result = re.findall(r"[abc]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result)
# ['a', 'b', 'a', 'a', 'a', 'c', 'a']
```

```python
# 3) 대괄호[] 안의 "-"는 범위를 나타낸다.
# - 문자0부터 문자9까지 [0123456789]를 줄여서 [0-9]로 표현
# - [a-z], [A-Z], ...
```

```python
import re

text = "A12B34"
print(re.findall(r"[0-9]", text))
# ['1', '2', '3', '4']
# 문자0부터 문자9까지 하나라도 일치하면 반환되는 리스트에 추가 
# 정규표현식 [0-9]는 [0123456789] 동일한 패턴
```

```python
# - 영문대문자 찾기
import re

text = "Python JAVA AI"
print(re.findall(r"[A-Z]", text))
```

```python
# 4) 문자제외 [^ ]
# - 대괄호 안에 ^는 괄호안에 문자를 제외한 문자
import re

text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
# 5) 아무 문자 .
# - . 은 줄바꿈을 제외한 아무 문자 한 글자를 의미한다.
import re

text = "cat ct cot cut coot"
result = re.findall(r"c.t", text) # "c_t" 이런 패턴의 문자(c로 시작, t로 끝나는 문자열길이 3) 찾기
print(result)
```

```python
# 6) 시작(^)과 끝($)
# - 문자열 안에서 패턴의 문자를 찾기보다는 문자열 전체가 특정 문자열로 시작 or 
# 끝나는 지를 확인할때 사용
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # ["Py"]

text = "python Python python"
print(re.findall(r"^Py", text)) # [] 
```

```python
# - search(정규표현식, text) 함수를 사용
# - 문자로 시작하면 match객체, 아니면 None 반환
import re

text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"^py", text)) # None
```

```python
import re

text = "Python"

print(re.search(r"hon$", text)) # <re.Match object; span=(3, 6), match='hon'>
print(re.search(r"abc$", text)) # None
```

```python
# 7) 수량자 *
# * 앞에 문자(한자)는 여러번 반복될수 있다
import re

text = "Pyyyyython"

# 시작이 Py 찾기
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'>
# 시작이 Pyyyyy.... 찾기
print(re.search(r"^Py*", text)) # <re.Match object; span=(0, 6), match='Pyyyyy'>
```

```python
# - match 함수를 사용하면 ^ 생략 가능
# - search 함수는 문자열 전체를 검색
# - match 함수는 문자열 시작만 검색하기 때문에 아래 3가지는 같은 결과 
print(re.search(r"^Py", text)) 
print(re.match(r"^Py", text))
print(re.match(r"Py", text))
```

```python
# 8) 수량자 ?
# - 앞 문자가 0개 또는 1개 존재한다.
# - "ab?c" : b가 없거나 1개 있거나 -> "abc"(O) 와 "ac"(O)는 매칭,  
# "abbc"(X) b가 2개라서 매칭 안됨

import re
words = [
    "ae",
    "abc",
    "abbcd",
    "abbbcde",
    "acde"
]

for word in words:
    print(re.match(r"ab?c", word))
# None
# <re.Match object; span=(0, 3), match='abc'>
# None
# None
# <re.Match object; span=(0, 2), match='ac'>
```

```python
# 9) \문자(백슬래시 문자)
import re

text = "Py@thon 123_xyz"

print(re.findall(r"\d", text))
print(re.findall(r"\D", text))
print(re.findall(r"\w", text))
print(re.findall(r"\W", text)) # 특수문자, 공백
print(re.findall(r"\s", text)) # 공백
```

```python
# 10) 수량자 {n}

# - 앞 문자가 정확히 n번 반복

import re
text = "1,23,456,7890,77a8888"
print(re.findall(r"\d{3}", text)) # 숫자문자가 3번연속 나오는 구간만 반환 리스트에 추가
# ['456', '789', '888']


text = "a bc def ghij kk@!#xx"
print(re.findall(r"\W{3}", text)) # 특수 문자가 3번연속
```

```python
# 11) 수량자 {n, m}
# 앞 문자가 최소 n번, 최대 m번 반복된다.

import re
text = "1 12 123 1234 12345"
print(re.findall(r"\d{2,4}", text)) # ['12', '123', '1234', '1234']
```

```python
# 12) OR( | ) 연산
# 여러 패턴 중 하나와 일치하면 된다.

import re
text = "cat, dog, bird, cats, dog"
print(re.findall(r"cat|dog", text)) # ['cat', 'dog', 'cat', 'dog']
```

```python
# 13) group()함수 매칭된 결과물을 반환
# - 괄호 ( )는 패턴을 하나의 그룹으로 묶는다.

import re

phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)
    if result != None:
        print(result)
        print(result.group())
```

```python
# 실습3) 이메일 검증

# ^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$
import re

emails = [
    "python@test.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
# 실습4) URL 검증

# ^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$
import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
# 실습5) 주민등록변호 형식 검사(외국사람 제외)

# ^\d{6}-[1-4]\d{6}$
# 숫자6시작 
# -문자 
# 1~4문자 하나 
# 숫자문자 6 끝
import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^\d{6}-[1-4]\d{6}$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
