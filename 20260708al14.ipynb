```python
local_sum = {}

for r in sales:
    price = r["판매액"]

    # price가 int가 아니면, 즉 결측치나 잘못된 값이면 건너뛰기
    if not isinstance(price, int):
        continue

    key_name = r["지역"]

    if key_name not in local_sum:
        local_sum[key_name] = 0

    local_sum[key_name] += price

for key, value in local_sum.items():
    print(f"{key}:{value}")
```

```python
#슬라이딩 윈도우에 사용될 리스트
prices=[]

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

#prices
window=3

#[0:3][1:4][2:5]...
for i in range(len(prices)-window+1):#range(len(prices)-window+1)
    window_prices=prices[i:i+window]
    avg=sum(window_prices)/window
    print(f'{i+1}~{i+window}일 이동평균: {avg}')
```

```python
#판매액의 누적합
prices=[]
for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# prices의 누적합 계산
cumulative_sum = []
total = 0
for p in prices:
    total += p
    cumulative_sum.append(total)
```

```python
print("판매액의 누적합:", cumulative_sum)
```

```python
#판매액의 평균 누적합
avgs=[]
for i in range(len(cumulative_sum)):
    avg = cumulative_sum[i] / (i + 1)
    avgs.append(avg)
print("판매액의 평균 누적합:", avgs)
```

```python
#이상치 탐지->처리
for r in sales:
    if isinstance(r['판매액'], (int)):
        if r['판매액'] >= 2000000:
            print("이상치 발견",r)
```

```python
#중앙값=값의 중앙에 위치한 값
#정렬 후 중앙값을 구한다
avg=sum(prices)/len(prices)
print("평균:", avg)

#중앙값
prices.sort()
print("정렬된 판매액:", prices)
if len(prices) % 2 == 1:
    median = prices[len(prices) // 2]
else:
    mid1 = prices[len(prices) // 2 - 1]
    mid2 = prices[len(prices) // 2]
    median = (mid1 + mid2) / 2
print("중앙값:", median)
```

```python
result = {}
for r in sales:
    p_name = r['상품']
    if p_name not in result:
        result[p_name] = {
            '합계': 0,
            '최대': 0,
            '최소': 0,
            '총수량': 0,
            '거래건수': 0
        }
        #{'노트북': {'합계': 0, '최대': 0, '최소': 0, '총수량': 0, '거래건수': 0},
        # '마우스': {'합계': 0, '최대': 0, '최소': 0, '총수량': 0, '거래건수': 0},
        # '키보드': {'합계': 0, '최대': 0, '최소': 0, '총수량': 0, '거래건수': 0},
        # '모니터': {'합계': 0, '최대': 0, '최소': 0, '총수량': 0, '거래건수': 0}}
       #이런식으로 추가할것이다.
    info=result[p_name]
    info['합계'] += r['판매액']
    info['총수량'] += r['수량']
    info['거래건수'] += 1
    #최대 최소
    if info['최대']< r['판매액']:
        info['최대']=r['판매액']
    if info['최소']==0 or info['최소']> r['판매액']:
        info['최소']=r['판매액']
print(result)
```
