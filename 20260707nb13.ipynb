```python
# 1. 누적합(Prefix Sum)
# 배열(리스트)의 특정 구간의 합을  빠르게 구하기 위해 사용하는 알고리즘 

# 만약 [10, 20, 30, 40, 50]이라는 배열이 있다고 가정해 보자.

# 이 배열에서 매번 특정 구간(예: 2번째부터 4번째까지)의 합을 구해야 한다면, 그때마다 더하기 연산을 반복해야 하므로 배열이 길어질수록 시간이 오래 걸린다.

# 이를 해결하기 위해 누적합 배열(Prefix Sum Array)을 미리 만들어 둔다.

# 누적합 배열(P)은 원본 배열(A) 사이즈+1 크기를 가진다. 계산의 편의를 위해 P[0] = 0 초기화 후 P[1]부터 원본배열의 누적합을 추가.

# -> 원본 A[0]까지의 누적합은 P[1], 

#     A[1]까지의 누적합은 P[2],

#     ... 

#     A[n]까지의 누적합은 P[n+1]에 저장된다.

# 원본 배열 (A): [10, 20, 30, 40, 50]
# 누적합 배열 (P): [0, 10, 30, 60, 100, 150] (계산 편의를 위해 맨 앞에 0을 추가합니다)
# ​
# 다음과 같은 리스트가 있다.
# 구간 합을 100번 물어본다면?

# 1) 첫번째 부터 끝까지의 합은?

# 2) 첫번째 부터 두번째까지의 합은?

# 3) 두번째 부터 4번째까지의 합은?

# 4) 세번째 부터 끝까지의 합은?
# ....
# 매번 반복문으로 원하는 구간의 누적을 계산해야 한다. 
# 앞 질문에서 누적합을 구한 구간도 다시 반복문으로 계산하여야 한다. -> 이런 반복되는 누적합의 질문이 예상되는 문제에 미리 전체 누적합을 계산해 두면 매번 반복문을 사용하지 않아도 구간합을 구할 수 있다. 
```

```python
nums = [10, 20, 30, 40, 50]

total = 0

# 특정 구간의 합을 물어볼때 마다 그 구간의 합을 구하는 연산을 반복하여야 한다.
for i in range(0, 4):
    total += nums[i]
print(total)

total = 0
# 이전에 물었던 구간과 반벅구간이 있어도 다시 연산을 반복하여야 한다.
for i in range(1, 3):
    total += nums[i]
print(total)
```

```python
nums = [10, 20, 30, 40, 50]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1] + n) # 누적합 값인 원소인 마지막 원소에 다음 원소값을 누적해서 append
    # 0+10->10 append [0, 10]
    # 10+20->30 append [0, 10, 30]
    # 30+30->60 append [0, 10, 30, 60]
    # ...  [0, 10, 30, 60, 100, 150]
print(prefix)
```

```python
# 3. 구간 합 공식
# 원본 배열 (A): [10, 20, 30, 40, 50]
# 누적합 배열 (P): [0, 10, 30, 60, 100, 150] (계산 편의를 위해 맨 앞에 0을 추가합니다)
# 원본배열 A

# 누적합배열 P

# 이렇게 누적합을 구해두면 이 누적합 배열(리스트)를 이용하여 구간 합 계산도 할 수 있다.

# A배열의 1번째부터 마지막 5번째까지(인덱스는 0~4 구간)의 구간합은? 배열 P[5]의 값 

# A배열의 2번째부터 마지막 5번째까지(인덱스는 1~4 구간)의 구간합은? 배열 P[5] - P[2-1]의 값

# A배열의 3번째부터 마지막 4번째까지(인덱스는 2~3 구간)의 구간합은? 배열 P[4] - P[3-1]의 값

# * 구간합 = P[end] - P[start-1] 라는 공식이 만들어 진다.
```

```python
nums = [10, 2, 30, 21 , 44, 5, 8, 50, 7]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1] + n)

# 주의 : 구간의 숫자는 인덱스가 아니다
# 3번째(값 30)에서 7번째(값 8)까지의 합
start = 3
end = 7

result = prefix[end] - prefix[start - 1]

print(result)
```

```python
# 1. 슬라이딩 윈도우(Sliding Window)
# 슬라이딩 윈도우(Sliding Window)는 배열이나 리스트에서 고정된 크기의 창(Window)을 옆으로 밀면서(Sliding) 문제를 해결하는 효율적인 알고리즘이다.

# start와 end 2개의 두 점 사이의 간격이 일정하게 유지되면서 이동하는 모습이 마치 창문을 이동시키는 모양새와 같다고 해서 슬라이딩 윈도우라는 이름으로 불리게 되었다.

# 다음과 같이 학생들의 점수 리스트가 있다면 사이즈=3의 윈도우를 이동시키면
```

```python
# scores = [10, 20, 30, 40, 50, 60]
# 3명씩(연속된) 총점을 구하면

# 10+20+30
# 20+30+40
# 30+40+50
# 40+50+60
# 반복 연산이 발생하게 된다
# 2. 슬라이딩 윈도우 개념
# 창(Window)의 크기가 3이면 다음과 같이 창이 이동한다.

# [10 20 30] 40 50 60     -> 윈도우 이동전에 합은 10+20+30=60 -> 리스트[0:창의크기]까지의 합  

# 창을 한 칸 이동하면     -> 이전합 + 새로추가된 숫자 - 이전 윈도우의 맨 앞 숫자  60+40-10
# 10 [20 30 40] 50 60

# 다시 창을 한 칸 이동하면 -> 90+50-20
# 10 20 [30 40 50] 60

# 다시 창을 한 칸 이동하면
# 10 20 30 [40 50 60]
# --> 창이 한 칸 이동하면 맨 앞 숫자는 빼고, 새로 들어온 숫자는 더한다.
```

```python
scores = [10, 20, 30, 40, 50, 60]

k = 3

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[:k])

print(window)

for i in range(k, len(scores)): # 원도우크기 부터 1칸식 이동하며 윈도우안의 합을 연산
    window = window - scores[i-k] + scores[i]
    print(window)
```

```python
# 학생이 100,000명이고 연속된 100명의 합을 구한다고 하면, 일반 반복문은 매번 100개를 다시 더하므로 매우 비효율적이다. 

# 슬라이딩 윈도우는 한 칸 이동할 때 앞의 값 1개를 빼고 뒤의 값 1개를 더하기만 하므로 훨씬 빠르다.

# 슬라이딩 윈도우는 "고정 크기의 구간을 한 칸씩 이동하면서 이전 계산 결과를 재사용하는 알고리즘"이다.
```

```python
# 3. 윈도우 구간 평균 구하기
scores = [80, 90, 70, 100, 95, 85]

k = 3

window = sum(scores[:k])

print(window / k)

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]
    print(window / k)
```

```python
#4. 윈도우 구간 합의 최대값(최대합) 구하기
scores = [10, 20, 30, 40, 50, 60]

k = 3

window = sum(scores[:k])

max_sum = window

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]

    if window > max_sum:
        max_sum = window

print(max_sum)
```

```python
# 5. 슬라이딩 윈도우는 어디 사용
# 이 알고리즘은 "연속적인 데이터 스트림"이나 "고정된 범위의 상태를 유지해야 할 때" 주로 사용 된다.

# - 네트워크 패킷 전송 (TCP Sliding Window): 네트워크에서 데이터를 보낼 때, 수신 측이 처리할 수 있는 만큼의 '윈도우 크기'를 정해두고, 데이터가 성공적으로 도착하면 윈도우를 옆으로 밀어가며 다음 데이터를 연속해서 전송한다. (흐름 제어의 핵심 기술)

# - 시계열 데이터의 이동평균(Moving Average) 계산: 주식 시장에서 '5일 이동평균선', '20일 이동평균선'을 구할 때 매일매일 지난 5일치 데이터를 새로 더하는 게 아니라, 슬라이딩 윈도우 방식으로 가장 오래된 날짜는 빼고 오늘 날짜는 더하며 빠르게 계산한다.

# - 문자열 매칭 및 패턴 탐색: 긴 문장 속에서 "길이 4짜리 특정 DNA 염기서열(예: AGCT)이 포함되어 있는가?"를 찾기 위해 글자 4개짜리 창문을 만들어 오른쪽으로 한 글자씩 밀어가며 탐색한다.

# - 실시간 데이터 스트리밍 분석: 최근 10분간의 실시간 트래픽 합계, 최근 1시간 동안의 웹사이트 방문자 수 변화 등을 계속 업데이트하며 모니터링 분석할 때 사용된다.
```

```python
totals = [window]
avgs=[window/k]
max_val=window
min_val=window

print(window)

for i in range(k, len(scores)):
    window = window - scores[i-k] + scores[i]
    totals.append(window)
    avgs.append(window/k)

    if window > max_val:
        max_val = window
    if window < min_val:
        min_val = window

print(totals)
print(avgs)
print(max_val)
print(min_val)
```

```python
# #3교시 트리
# 데이터 분석가는 트리(Tree) 알고리즘을 직접 구현하는 경우는 드물다. 대신, 트리로 구현된 머신러닝 라이브러리(예: Decision Tree, Random Forest, XGBoost)를 사용해 데이터를 분석하고 예측하는 코드를 자주 작성한다.

# 라이브러리를 사용하면 fit()이나 predict() 같은 함수 한 줄로 트리를 사용할 수 있지만, 내부에서 어떤 원리로 결과가 만들어지는지 이해하지 못하면 모델을 올바르게 활용하거나 결과를 해석하기 어렵다.

# 따라서 트리의 기본 구조와 탐색 원리를 이해하면 라이브러리가 내부적으로 수많은 조건문(if-else)을 트리 형태로 생성하고, 그 트리를 따라가며 예측을 수행한다는 것을 쉽게 이해할 수 있다.
```

```python
# 1. 트리 자료구조

# ​

# 1) 트리 자료구조의 목적

# ​

# 계층적인 데이터들을 빠르게 탐색/검색/관리가 가능하한 자료구조

# 트리(이진탐색트리)는 삽입과 삭제에 탐색 과정이 필요하지만(저장할 위치를 찾는 과정이 필요해서 리스트보다 비용이 더 드는 경우가 많다), 데이터를 빠르게 탐색하기 위해 사용하는 자료구조이다.

# ​

# * 선형 구조 vs 트리 구조

# ​

# - 선형 구조: 앞에서부터 순서대로 탐색

# - 트리 구조: 부모-자식 구조를 따라 필요한 데이터만 탐색하여 더 빠르게 찾을 수 있음. 물론 트리가 한쪽으로 치우치면 최악의 경우 선형구조와 동일한 탐색시간이 느려질 수 있다.

# ​
```

```python
#- 선형 구조
nums = [10, 20, 30, 40, 50]

# 40 찾기
for n in nums:
    if n == 40:
        print("찾음")
#[10] → [20] → [30] → [40] → [50]
#                   ↑
#         순서대로 하나씩 탐색
# - 트리 구조

# # 예시 트리
#         30
#        /  \
#      20    40
#     /        \
#   10          50

# # 40 찾기
# # 30 → 40 (바로 탐색)
#         30
#        /  \
#      20    40 ← 찾음
#     /        \
#   10          50
```

```python
# 2. if문을 Tree로 표현
# - if문
color = "Red"
size = "Large"

if color == "Red":
    if size == "Large":
        print("사과 (Apple)")
    else:
        print("체리 (Cherry)")
elif color == "Purple":
    print("포도 (Grape)")
else:
    print("모르는 과일")
```

```python
# - Tree
# 이때 트리 자료구조는 데이터를 저장하기 위한 목적보다는 조건 업무프로세스의 순서를 저장
fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}
```

```python
# - 트리모양

# 조건 분기 프로세스를 하나의 트리 자료구조형태로 저장

#                  color == "Red" ?

#                  /             \
#               Yes              No
#                |                |
#         size == "Large"?    color=="Purple"?
#            /      \            /         \
#         Yes      No         Yes         No
#         |         |          |           |
#       사과      체리       포도      모르는 과일
```

```python
# 3. 트리 탐색 코드
# isinstance(a,b)
# ->bool 함수는 변수 a가 b타입의 객체(instance)이면 True 아니면 False를 반환하는 함수
```

```python
fruit = {
    "color": "Red",
    "size": "Large" # Small
}

fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}

node = fruit_tree #node변수의 초기값은 dict-> 무한루프 ->node변수값이 dict아닐때까지

# 트리를 탐색하는 반복문
while isinstance(node, dict): # node객체가 dict 이라면 True 아니면 False

    # dict이라면 "question"키값을 찾아 질문은 key, 정답은 value에 저장 
    key, value = node["question"] 
   
    # 입력데이터의 key값과 비교하여 정답이면 "yes"키값을 node에 대입, 아니면 "no"키값을 node에 대입
    # 이렇게 대입된 node값이 dict 타입이면 다시 반복문 진행... 
    # 이렇게 node에 dict타입이 아닌 타입(위 트리에선 문자열)이 올때 까지 반복문은 계속
    if fruit[key] == value:
        node = node["yes"]
    else:
        node = node["no"]

# 반복문을 빠져 나오면 마지막 node값을 출력
print(node)
```
