```python
# 1. 클래스
# 클래스는 사용자 정의 데이터 타입이다.
# 클래스는 객체를 만들기 위한 설계도(틀) 이다(객체지향 프로그래밍(OOP)의 핵심 개념)
# * 기존의 딕셔너리 타입에 객체지향 패러다임을 추가한 데이터 타입 

# 2. 클래스 기본 구조
# class : 클래스 정의 키워드
# __init__ : 생성자 (객체 생성 시 자동 실행)
# self : 자기 자신 객체를 의미(자바의 this와 동일)
# def : 메서드 정의

# class 클래스이름:
#     def __init__(self, 매개변수):
#         self.속성 = 값

#     def 메서드이름(self):
#         실행코드
```

```python
# * 메서드(멤버함수) vs 함수
# - 메서드는 특정 클래스로 만든 하나의 객체(데이터)에만 종속되는 기능으로 클래스 소스 코드안에 같이 정의 되어 있어야한다.
# - 함수는 특정 클래스로 만든 모든 객체(데이터)에 사용가능한 기능으로 클래스와 별개로 기능을 사용하는 곳에 함수가 있어야 한다.

# ex) 
# 메서드 : 하나의 TV에만 사용가능한 리모콘으로 TV를 구매하면 포함되어 있으며 이 TV에만 (종속)사용 가능하다.
# 함수 : TV와는 별개로 구매해야 하고, 같은 종류의 모든 TV에 사용 가능하다.
# 어떤 클래스 데이터타입의 객체에 종속되는 기능은 멤버함수(메서드)로 정의하고, 어떤 데이터타입에도 종속되지 않는 기능은 함수로 정의하면 된다.
```

```python
class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2

    def print_info(self):
        print(f"이름 : {self.name}")
        print(f"국어 : {self.kor}")
        print(f"영어 : {self.eng}")
        print(f"총점 : {self.total()}")
        print(f"평균 : {self.average()}")


s1 = Student("홍길동", 90, 85)
s2 = Student("이순신", 100, 95)

s1.print_info()
print()
```

```python
class Account:

    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, money):
        self.balance += money

    def withdraw(self, money):
        self.balance -= money

    def print_info(self):
        print(f"예금주 : {self.owner}")
        print(f"잔액 : {self.balance}원")


acc = Account("홍길동", 10000)

acc.deposit(5000)
acc.withdraw(3000)

acc.print_info()
```

```python
class Car:

    def __init__(self, model):
        self.model = model
        self.speed = 0

    def accelerate(self):
        self.speed += 10

    def brake(self):
        self.speed -= 10

    def print_info(self):
        print(f"차량 : {self.model}")
        print(f"속도 : {self.speed}km/h")


car = Car("소나타")

car.accelerate()
car.accelerate()
car.brake()

car.print_info()
```

```python
class TV:

    def __init__(self):
        self.power = False
        self.channel = 1

    def on(self):
        self.power = True

    def channel_up(self):
        self.channel += 1

    def print_info(self):
        print(f"전원 : {self.power}")
        print(f"채널 : {self.channel}")


tv = TV()

tv.on()
tv.channel_up()
tv.channel_up()

tv.print_info()
```

```python
class Rectangle:

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return (self.width + self.height) * 2


r = Rectangle(10, 5)

print(r.area())
print(r.perimeter())
```

```python
# 1. 객체지향 패러다임
# - 앞서 클래스는 기존의 딕셔너리 타입에 객체지향 패러다임을 추가한 데이터 타입이다. 딕셔너리 vs 클래스 예제 중심으로 설명했다
# - 딕셔너리는 데이터를 저장하는 자료구조이고, 클래스는 데이터를 저장하는 속성과 그 데이터를 처리하는 메서드를 하나로 묶어 객체를 만드는 설계도이다. 
# 이러한 객체를 중심으로 프로그램을 설계하는 방식이 객체지향 프로그래밍(Object-Oriented Programming, OOP)이며, 이러한 사고방식을 객체지향 패러다임이라고 한다. 
# - 객체지향 프로그래밍(Object-Oriented Programming, OOP) 특징

# ※ 다음의 특징들은 파이썬 OOP에서 수업
# - 객체(Object) 중심으로 프로그램을 설계한다.
# 데이터가 아닌 '학생', '자동차', '은행계좌' 같은 객체를 중심으로 생각한다. 
# - 데이터와 기능을 하나로 묶는다. (캡슐화)
# 속성(데이터)과 메서드(기능)를 하나의 객체로 관리한다. 
# - 코드 재사용이 쉽다.
# 클래스를 여러 번 생성(객체 생성)하여 사용할 수 있다. 
# - 상속(Inheritance)이 가능하다.
# 기존 클래스를 물려받아 새로운 클래스를 만들 수 있다. 
```

```python
# 상속의 이해를 돕기위한 간단한 예)
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)

# 상속하면
class HighStudent(Student):
    def study(self):
        print("공부한다.")

# 사용
s = HighStudent("홍길동")
s.print_name()
s.study()
```

```python
# - 유지보수가 쉽다.
# 기능이 변경되면 해당 클래스만 수정하면 되므로 수정 범위가 줄어든다. 
# - 확장이 쉽다.
# 새로운 기능이나 클래스를 기존 코드에 큰 영향을 주지 않고 추가할 수 있다. 
# - 현실 세계를 모델링하기 쉽다.
# 사람, 자동차, 학생, 동물처럼 실제 사물을 객체로 표현하기에 적합하다.
```

```python
# 2. 딕셔너리의 함수 vs 클래스의 메서드
# 딕셔너리도 함수(객체)를 값으로 저장할 수 있지만, 
# 데이터와 함수는 서로 독립적인 객체이며 자동으로 연결되어 있지 않다. 
# 반면 클래스에서는 데이터와 메서드가 하나의 객체에 묶여 함께 동작한다.
```

```python
# 1) 딕셔너리 함수
def average(student):
    return (student["kor"] + student["eng"]) / 2


student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "average": average
}

print(student["average"](student))
# - 데이터 객체(name, kor, eng), 함수 객체(average)가 각각 따로 존재하고, 
# 딕셔너리는 단순히 둘을 저장하고 있을 뿐이다.
# - 함수에는 학생에 대한 정보가 없다.(누구의 평균을 계산하는지 모른다) 
# 그래서 호출할때 학생 인수로 학생의 정보를 전달해야만 한다.
```

```python
# 2) 클래스 메서드
# - 메서드는 자기가 다루어야 할 학생의 정보(자신의 객체 : self)를 이미 알고 있다.
class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def average(self):
        return (self.kor + self.eng) / 2


s = Student("홍길동", 90, 85)

print(s.average())

# 메서드는 객체 내부에 소속되어 있으며, 자신의 데이터(self)를 직접 사용할 수 있다.
# 딕셔너리도 함수를 값으로 저장할 수 있지만, 함수는 딕셔너리에 소속된 것이 아니라 독립적인 객체이다. 따라서 데이터를 항상 인수로 전달해야 한다. 
# 반면 클래스의 메서드는 객체에 소속되어 self를 통해 자신의 데이터를 직접 사용할 수 있다는 점이 객체지향의 핵심 차이이다.
```

```python
# 3. 딕셔너리 사용 vs 클래스 사용

# 1) 딕셔너리만 사용
student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85
}

print(student["name"])
print(student["kor"] + student["eng"])
```

```python
# 2) 클래스로 변경하면
class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2


s1 = Student("홍길동", 90, 85)
print(s1.average())

s2 = Student("김철수", 80, 88)
print(s2.average())
```

```python
# - "객체지향 프로그래밍(OOP)은 프로그램을 함수 중심이 아니라 '객체(Object)' 중심으로 만드는 방법"이다.
# 1. 절차지향 프로그래밍
# 절차(순서)대로 코드를 작성한다. -> 데이터를 만들고 그 데이터를 사용(참조)하는 함수를 만들어 호출한다. 
# -> 데이터와 함수가 서로 따로 존재 한다.
student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85
}

total = student["kor"] + student["eng"]
avg = total / 2

print(total)
print(avg)
# 데이터
#     ↓
# 함수 실행
#     ↓
# 결과 출력
```

```python
# 2. 객체지향 프로그래밍
class Student:

    def __init__(self, name, kor, eng):
        self.name = name
        self.kor = kor
        self.eng = eng

    def total(self):
        return self.kor + self.eng

    def average(self):
        return self.total() / 2


s = Student("홍길동", 90, 85)

print(s.total())
print(s.average())
# - 학생 객체안에 자가 자신의 객체(self)를 다루는 메서드 포함 -> 데이터와 기능을 하나로 묶는다. 
# 학생 객체
#  ├── 이름
#  ├── 국어
#  ├── 영어
#  ├── 총점 계산()
#  └── 평균 계산()
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_sales(self, quantity):
        return self.price * quantity

    def print_info(self):
        print(f"상품명: {self.name}")
        print(f"가격: {self.price}원")
        print(f"재고: {self.stock}개")
```

```python
p = Product("키보드", 50000, 10)

print(p.name)
print(p.price)
print(p.stock)
```

```python
products=[
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3)
]
print(products)
```

```python
for p in products:
    p.print_info()
    print()
```

```python
for p in products:
    if p.stock >= 10:
        p.print_info()
        print()
```

```python
#salses index
sales=[
    Product("키보드", 50000, 10),
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3),
    Product("노트북", 1200000, 3),
    Product("노트북", 1200000, 3)
]

#이런 세일즈 목록이 있을때 구해야된다.
#{'키보드':2,'마우스':1,'모니터':1,'노트북':3}이런 결과 나오게
counts = {}
for s in sales:
    if s.name in counts:
        counts[s.name] += 1
    else:
        counts[s.name] = 1
print(counts)
```

```python
check_list = []#같은 상품이름 2번 계산되면 안되니까 계산했으면 입력
for s in sales:
    if s.name not in check_list:
        print(f"{s.name} : {s.total_sales(counts[s.name])}원")
        check_list.append(s.name)
```
