```python
# 1. 리스트의 특징과 활용

# 리스트는 맨 뒤 데이터 추가(append()),  맨 뒤 데이터 삭제(pop()) 가 매우 빠르다. 그래서 스택(Stack) 을 구현하기에 적합하다.

# 2. 스택(Stack)

# LIFO(Last In First Out)는 나중에 들어온 데이터가 가장 먼저 나온다.

#        ┌───┐
# Top →  │ C │
#        ├───┤
#        │ B │
#        ├───┤
# Bottom │ A │
#        └───┘
# - 스택 연산

# push : 데이터 넣기 -> 리스트의 append() 사용

# pop : 데이터 거내기 -> 리스트의 pop() 사용
```

```python
stack = []

stack.append("접시1")
stack.append("접시2")
stack.append("접시3")

while len(stack) > 0:
    print(stack.pop())
```

```python
plate=[]
plate.append({'color':'red', 'price':'100'})
plate.append({'color':'yellow', 'price':'150'})
plate.append({'color':'blue', 'price':'200'})
plate.append({'color':'black', 'price':'250'})

print(plate)

total_price=0
while len(plate)>0:
    p=plate.pop()
    total_price+=int(p['price'])
print(total_price)
```

```python
# ※ 큐를 만들때 원소가 많으면 리스트를 잘 사용하지 않는다.

# queue = []

# queue.append(10)
# queue.append(20)
# queue.append(30)
# queue.append(40)
# queue.append(50)
# queue.pop(0)
# 위 코드는 큐처럼 동작하지만, pop(0)을 실행하면 첫 번째 원소를 삭제한 뒤 나머지 원소를 모두 한 칸씩 앞으로 이동해야 한다.

# 삭제 전
# [10, 20, 30, 40, 50]

# pop(0)

# 삭제 후
# [20, 30, 40, 50]
# 20, 30, 40, 50이 모두 앞으로 이동하므로 데이터가 많을수록 비효율적이다.

# 그래서 실제 알고리즘에서는 큐를 구현할 때 collections모듈의 deque를 사용한다.

# deque의 popleft() 메서드를 사용하면 맨 앞의 데이터를 이동 없이 빠르게 꺼낼 수 있어 알고리즘 문제에서는 deque를 사용하는 것이 표준적인 방법이다. deque는 다음 시간에서  상세히 다룬다.
```

```python
from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue.popleft())
print(queue.popleft())
```

```python
data=[]
count=1
for i in range(3):
    row=[]
    for j in range(3):
        row.append(count*10)
        count+=1
    data.append(row)
print(data)

for row in data:
    for val in row:
        print(row,val,end=',')
```

```python
#게임 또는 수학에서 2차원리스트 좌표가 필요한 경우
for i in range(len(data)*len(data[0])):
    print(data[i//len(data[0])][i%len(data[0])], end=',')
#i=0일때 [0,0], i=1일때 [0.1], i=2일때 [0,2]
#i=3일때 [1,0], i=4일때 [1,1], i=5일때 [1,2].....
#앞좌표는 3(len(data[0]))의 사이즈의 몫
#뒤 좌표는 3의 나머지
```

```python
for i in range(len(data) * len(data[0])):
    row = i // len(data[0])
    col = i % len(data[0])

    print(f'[{row},{col}] = {data[row][col]}')
```

```python
for i in range(len(data) * len(data[0])):
    print(data[i // len(data[0])][i % len(data[0])], end=',')

    if i % len(data[0]) == len(data[0]) - 1:
        print()
```

```python
from collections import deque
dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)

dq = deque()
dq.append(20)
dq.append(30)
dq.appendleft(10)

print(dq)
```

```python
dq = deque([10, 20, 30])
x = dq.pop()

print(x)
print(dq)

dq = deque([10, 20, 30])
x = dq.popleft()

print(x)
print(dq)
```

```python
dq = deque()

dq.append(10)
dq.append(20)
dq.appendleft(5)
dq.appendleft(1)

print(dq)

print(dq.pop())
print(dq.popleft())

print(dq)
```

```python
dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)

print(dq) # deque([5, 1, 2, 3, 4])
```

```python
dq = deque([1, 2, 3, 4, 5])
dq.rotate(-2) # 왼쪽으로 두 칸 회전

print(dq)
```

```python
from collections import deque

history = deque(maxlen=5) # maxlen=n n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

```python
from collections import deque

cards = deque([1, 2, 3, 4, 5])

while len(cards) > 1:
    print("버림 :", cards.popleft())
    cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
